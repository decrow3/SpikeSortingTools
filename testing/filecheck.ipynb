```python
import numpy as np
from pathlib import Path

# Paths to both timestamps files
path1 = Path('/mnt/NPX/Gru/20211217/2021-12-17_13-16-50/Record Node 101/experiment1/recording1/continuous/Neuropix-PXI-100.0/timestamps.npy')
path2 = Path('/mnt/NPX/Gru/20220412/2022-04-12_12-42-59/Record Node 101/experiment1/recording1/continuous/Neuropix-PXI-100.0/timestamps.npy')

def check_timestamps(path):
    if path.is_file():
        arr = np.load(str(path), mmap_mode='r')
        print(f"{path}:")
        print(f"  Shape: {arr.shape}")
        print(f"  Dtype: {arr.dtype}")
        print(f"  First 5: {arr[:5]}")
        print(f"  Last 5: {arr[-5:]}")
        print(f"  Min: {arr.min()}, Max: {arr.max()}")
        print(f"  Is monotonically increasing: {np.all(np.diff(arr) >= 0)}")
        print()
    else:
        print(f"{path} not found.\n")

check_timestamps(path1)
check_timestamps(path2)
```
